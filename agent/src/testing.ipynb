```python
import pandas as pd
import numpy as np
import pickle
```

```python
vector = pickle.load(open("./vectorizer.pkl", 'rb'))
model = pickle.load(open("./model.pkl", 'rb'))
```

```python
df = pd.read_csv("../Dataset/test.csv")
```

```python
df.head()
```

```python
ans = {
    "bad": 0,
    "good": 0
}

for i in range(len(df)):
    url = df['url'][i]
    result = model.predict(vector.transform([url]))[0]
    
    if result == 'bad':
        ans['bad'] += 1
    elif result == 'good':
        ans['good'] += 1
```

```python
ans
```

```python
accuracy = (ans['bad'] / (ans['bad'] + ans['good'])) * 100
accuracy
```
